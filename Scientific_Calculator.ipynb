```python
import math
import re

# Safe environment to evaluate expressions
safe_dict = {
    'sin': math.sin,
    'cos': math.cos,
    'tan': math.tan,
    'asin': math.asin,
    'acos': math.acos,
    'atan': math.atan,
    'sqrt': math.sqrt,
    'log': math.log,       # natural log
    'log10': math.log10,
    'ln': math.log,
    'exp': math.exp,
    'pow': math.pow,
    'abs': abs,
    'round': round,
    'floor': math.floor,
    'ceil': math.ceil,
    'factorial': math.factorial,
    'pi': math.pi,
    'e': math.e
}

# For storing memory/variables
variables = {}

# History
history = []

def assign_variable(expr):
    match = re.match(r"([a-zA-Z_]\w*)\s*=\s*(.+)", expr)
    if match:
        var, val_expr = match.groups()
        try:
            val = eval(val_expr, {"__builtins__": None}, {**safe_dict, **variables})
            variables[var] = val
            print(f"{var} = {val}")
            history.append(f"{var} = {val}")
        except Exception as e:
            print(f"Error: {e}")
        return True
    return False

def calculate(expr):
    try:
        result = eval(expr, {"__builtins__": None}, {**safe_dict, **variables})
        print("Result:", result)
        history.append(f"{expr} = {result}")
    except Exception as e:
        print("Error:", e)

# Main loop
print("Advanced Scientific Calculator")
print("Type expressions or variable assignments (e.g., x = 5)")
print("Use math functions: sin(), cos(), log(), factorial(), etc.")
print("Type 'history' to view previous calculations.")
print("Type 'vars' to view saved variables.")
print("Type 'exit' to quit.\n")

while True:
    expr = input(">>> ")
    if expr.lower() == 'exit':
        break
    elif expr.lower() == 'history':
        print("\n".join(history))
    elif expr.lower() == 'vars':
        print(variables)
    elif assign_variable(expr):
        continue
    else:
        calculate(expr)
```
